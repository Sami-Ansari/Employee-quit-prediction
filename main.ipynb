# Important Features

```python
features_imp = pd.DataFrame(dt.feature_importances_,index = X_train.columns, columns =["Importance"]).sort_values('Importance',ascending= False)
```

```python
features_imp
```

```python
def plot_feature_importance(model):
    plt.figure(figsize=(10, 10))
    n_features =9
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features),X) # np fro 9 values and X for name of values
    plt.xlabel('Feature Importance')
    plt.ylabel('Features')
    plt.ylim(-1, n_features)
    
plot_feature_importance(dt)
```

# KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train, y_train)
print(knn.score(X_test, y_test))
```

```python
predictions_knn = knn.predict(X_test)
```

```python
accuracy_score_knn = accuracy_score(y_test, predictions_knn)*100
```

```python
accuracy_score_knn
```

```python
k_range = range(1,26)
score = {}
score_list = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    predictions = knn.predict(X_test)
    score[k] = accuracy_score(y_test, predictions)
    score_list.append(accuracy_score(y_test, predictions))
```

```python
score
```

```python
plt.plot(k_range,score_list)
```
